# **Analisando o número de Imigrantes para o Canadá com um DataFrame e criando gráficos e figuras.**

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/imigrantes_canada.csv')
```

```python
df
```

```python
df.info()
```

```python
#mudando o index para país
df.set_index('País', inplace=True)
```

```python
#Criando uma variável para armazenar o intervalo de tempo, com o map aplicando isto em todo o intervalo.
#2014 pois o range não conta o último valor
anos = list(map(str, range(1980, 2014)))
```

```python
anos
```

```python
#essa propriedade consegue pegar rótulos específicos dentro de um dataframe.
brasil = df.loc['Brasil', anos]
```

```python
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
```

```python
dados_brasil
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4)) #modificando a altura do gráfico
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.title('Imigração do Brasil para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Número de Imigrantes')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010']) #definindo a marcação do eixo x para melhor visualização do gráfico
plt.show()
```

# Criando uma figura desse gráfico
**Motivos:** Mais fácil de modificar e possui mais ferramentas de otimização

```python
fig, ax = plt.subplots(figsize=(8,4)) #fig, cria uma figura. ax, permite modificar os eixos.
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
ax.set_title('Imigração do Brasil para o Canadá\n1980-2013')
ax.set_xlabel('Ano')
ax.set_ylabel('Número de Imigrantes')
ax.xaxis.set_major_locator(plt.MultipleLocator(5)) #função para modificar este eixo y. com intervalo de 5 em 5 anos.
plt.show()
```

# Comparação entre gráfico de linhas e boxplot
O boxplot se trata de extremidades, essas extremidades representam a amplitude, ou seja, o valor mínimo na parte inferior e o valor máximo na parte superior. No exemplo, o valor mínimo é de aproximadamente 100 pessoas, já o valor máximo está entre 1500 a 2000. A linha inferior mínima indica os 25% dos anos, a linha laranja central indica 50% e a linha máxima indica que em 75% dos anos, o número de imigrantes foi cerca de 1000.

```python
fig,axs = plt.subplots(1, 2, figsize=(15,5)) #indica o número de linhas(1) e o número de figuras(2).
axs[0].plot(dados_brasil['ano'], dados_brasil['imigrantes']) #índice 0 indica para o matplot desenhar o primeiro subplot(gráfico de linhas)
axs[0].set_title('Imigração do Brasil para o Canadá\n1980-2013')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Número de Imigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5)) #para frequência de 5 anos
axs[0].grid() #adiciona uma grade no gráfico


axs[1].boxplot(dados_brasil['imigrantes']) #índice 1 indica para o matplot desenhar o segundo subplot
axs[1].set_title('Boxplot da Imigração do Brasil para o Canadá\n1980-2013 ')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Número de Imigrantes')
axs[1].grid() #adiciona uma grade no gráfico

plt.show()
```

# Comparação do número de imigrantes do Brasil e Argentina no Canadá:

```python
df_comparacao = df.loc[['Brasil', 'Argentina'], anos] #extraindo os dados dos dois países
df_comparacao
```

```python
df_comparacao = df_comparacao.T #transposição da matriz, linhas por colunas
df_comparacao
```

```python
plt.plot(df_comparacao['Brasil'],label = 'Brasil') #plotando o gráfico e adicionando rótulos com o label
plt.plot(df_comparacao['Argentina'], label = 'Argentina')
plt.title('Imigração do Brasil e Argentina para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Número de Imigrantes')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.legend() #adiciona uma legenda ao gráfico
plt.show()
```

Conclusão: a partir dos anos 2000, os números de imigrantes na **Argentina** começaram a aumentar e atingiram os valores mais elevados entre **2000** e **2005**, antes de cair e atingir valores muito baixos nos anos mais recentes. Por outro lado, no caso do **Brasil**, os aumentos na imigração começaram após **2005** e parecem ter atingido o valor mais alto em **2010**.

# Tendências de imigração para os países da América do Sul

```python
fig, axs = plt.subplots(2,2, figsize=(10,6))
fig.subplots_adjust(hspace=0.5, wspace=0.3) #ajustando o espaçamento horizontal e vertical
fig.suptitle('Imigração dos quatro maiores países da América do Sul para o Canadá de 1980 a 2013')

axs[0,0].plot(df.loc['Brasil', anos])
axs[0,0].set_title('Brasil')

axs[0,1].plot(df.loc['Colômbia', anos])
axs[0,1].set_title('Colômbia')

axs[1,0].plot(df.loc['Peru', anos])
axs[1,0].set_title('Peru')

axs[1,1].plot(df.loc['Venezuela', anos])
axs[1,1].set_title('Venezuela')

for ax in axs.flat: #percorre toda a matriz e eaplica em todos os 4 gráficos
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.set_ylabel('Número de Imigrantes')
  ax.set_xlabel('Ano')
  ax.grid()

ymin = 0
ymax = 7000

for ax in axs.ravel(): #método para definir um parâmetro de valores ao eixo y
 ax.set_ylim(ymin, ymax)
plt.show()
```
